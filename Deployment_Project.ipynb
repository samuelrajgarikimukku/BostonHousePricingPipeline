# Lets load the Boston House Pricing Dataset

```python
from sklearn.datasets import load_boston
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
dataset= pd.DataFrame(data)
display(dataset.head())
```

```python
dataset.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
```

```python
dataset['Price'] =target
```

```python
## Independent and Dependent features

X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
##  Standardize the dataset StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
import pickle
pickle.dump(scaler,open('scaling.pkl','wb'))
```

```python
X_train
```

# Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
## Print the coefficients and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
## on which parameters this model has been trained
regression.get_params()
```

```python
### Prediction With Test Data
y_pred = regression.predict(X_test)
```

```python
y_pred
```

### Predictions Plots

```python
## Plot a scatter plot for prediction

plt.scatter(y_test,y_pred)
```

```python
## Residuals
residuals = y_test - y_pred
```

```python
residuals
```

```python
## Plot this residuals

sns.displot(residuals,kind="kde")
```

```python
## Scatter plot with respect to prediction and residuals
## Unifrom distribution
plt.scatter(y_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(np.sqrt(mean_squared_error(y_test,y_pred)))
```

## R Square and adjusted R Square

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test,y_pred)
print(score)
```

```python
'''There is no direct library for adjusted R-Square so we have to directly
implemet it '''

1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
